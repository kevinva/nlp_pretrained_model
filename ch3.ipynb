````python
from ltp import LTP
````

````python
ltp = LTP()
````

````python
segment, hidden = ltp.seg(['南京市长江大桥。'])
print(segment)
````

````python
sentences = ltp.sent_split(['南京市长江大桥。', '汤姆生病了😷。他去了医院。'])
print(sentences)

segment, hidden = ltp.seg(sentences)
print(f'segment: {segment}')
print(f'hidden: {hidden}')
````

````python
pos_tags = ltp.pos(hidden)
print(pos_tags)
````
